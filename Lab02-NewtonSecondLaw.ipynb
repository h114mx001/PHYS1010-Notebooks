```python
accelerations = [0.25, 0.49, 0.71, 0.1237, 0.1163, 0.1150]
times = [3.56, 2.23, 2.32, 4.27, 4.39, 4.36]
f(a) = 1/2*a*x^2
curves = [f(a) for a in accelerations]
l = len(accelerations)
colors = rainbow(l)
sum(plot(curves[k], [x, 0, times[k]], color=colors[k]) for k in range(0, l)).show()
```

```python
v(a) = a*x
vCurves = [v(a) for a in accelerations]
sum(plot(vCurves[k], [x, 0, times[k]], color=colors[k]) for k in range(0, l)).show()
```
